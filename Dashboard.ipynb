```python
# python program demonstrating 
# Combobox widget using tkinter 
import tkinter as tk 
from tkinter import *
from tkinter.ttk import *
from tkinter import ttk 
import pandas as pd
import import_ipynb
from tkinter.filedialog import askopenfile 
import Preprocess_Summary
from Preprocess_Summary import preprocess
from Preprocess_Summary import summary
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline

# Creating tkinter window 
window = tk.Tk() 
window.title('Information Extraction') 
window.geometry('1000x850') 

# label text for title 
ttk.Label(window, text = "Information Extraction Dashboard", background = 'green', foreground ="white", 
          font = ("Times New Roman", 15)).grid(row = 0, column = 1) 

# label 
ttk.Label(window, text = "Select the Session :", font = ("Times New Roman", 10)).grid(column = 0, row = 5, padx = 10, pady = 25) 

# Combobox creation 
n = tk.StringVar() 
country = ttk.Combobox(window, width = 27, values=("India","Brazil","Australia")) 

country.grid(column = 1, row = 5) 
country.current() 

# label 
ttk.Label(window, text = "Select Output Type :",  font = ("Times New Roman", 10)).grid(column = 0,  row = 7, padx = 10, pady = 25) 

# Combobox creation 
n = tk.StringVar() 
file_type = ttk.Combobox(window, width = 27, values=("File","Summary","KG")) 

file_type.grid(column = 1, row = 7) 
file_type.current() 

ttk.Label(window, text = "Output :", font = ("Times New Roman", 10)).grid(column = 0,  row = 14, padx = 10, pady = 25) 

T = tk.Text(window, height = 30, width = 102)
T.grid(column=1,row=18)
scroll = tk.Scrollbar(window)
scroll.grid(row=18,column=2,padx=12,sticky='NS')
scroll.config(command=T.yview)
 

    
def checkcmbo():
    print(country.get())
    print(file_type.get())
    T.delete(1.0,tk.END)
    if country.get() == "India"  :
        if file_type.get()=="File":
            file=open("Converted sessions\Session 25 - 1970\IND_25_1970.txt")
            res=file.read()
            T.insert(tk.END, res)
        elif file_type.get()=="Summary":
            input_string=open("Converted sessions\Session 25 - 1970\IND_25_1970.txt")
            res=summary(input_string)
            T.insert(tk.END,res)
        elif file_type.get()=="KG":
            file=open("Converted sessions\Session 25 - 1970\IND_25_1970.txt")
            text=file.read()
            df=pd.DataFrame()
            df=preprocess(text)
            img = tk.PhotoImage(file = "graph_new.png")
            #print(img)
            def add_image():
                    T.image_create(tk.END, image = img) # Example 1
                    #print("img found")
            tk.Button(window, text = "Insert img", command = add_image)
            add_image()
            
    elif country.get() == "Brazil"  :
        if file_type.get()=="File":
            file=open("Converted sessions\Session 25 - 1970\BRA_25_1970.txt")
            res=file.read()
            T.insert(tk.END, res)
        elif file_type.get()=="Summary":
            input_string=open("Converted sessions\Session 25 - 1970\BRA_25_1970.txt")
            res=summary(input_string)
            T.insert(tk.END,res)
        elif file_type.get()=="KG":
            file=open("Converted sessions\Session 25 - 1970\BRA_25_1970.txt")
            text=file.read()
            df=pd.DataFrame()
            df=preprocess(text)
            img = tk.PhotoImage(file = "graph_new.png")
            def add_image():
                    T.image_create(tk.END, image = img) # Example 1
            tk.Button(window, text = "Insert img", command = add_image)
            add_image()
            
    elif country.get() == "Australia"  :
        if file_type.get()=="File":
            file=open("Converted sessions\Session 25 - 1970\AUS_25_1970.txt")
            res=file.read()
            T.insert(tk.END, res)
        elif file_type.get()=="Summary":
            input_string=open("Converted sessions\Session 25 - 1970\AUS_25_1970.txt")
            res=summary(input_string)
            T.insert(tk.END,res)
        elif file_type.get()=="KG":
            file=open("Converted sessions\Session 25 - 1970\AUS_25_1970.txt")
            text=file.read()
            df=pd.DataFrame()
            df=preprocess(text)
            img = tk.PhotoImage(file = "graph_new.png")
            def add_image():
                    T.image_create(tk.END, image = img) # Example 1
            tk.Button(window, text = "Insert img", command = add_image)
            add_image()

def close_window(): 
    window.destroy() 
    
        
btn = tk.Button(window, text = 'show output!', bd = '5', 
                          command = checkcmbo)         #T.insert(END,'res')
btn.grid(column=1,row=10)

btn1=tk.Button(window,text="Quit",bd='5',command=close_window)
btn1.grid(column=1,row=14)

window.mainloop() 

```
